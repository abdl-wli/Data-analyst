```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('survey_results_public.csv',index_col=['ResponseId'])
pd.set_option('display.max_columns',84)
data
```

```python
pak_devs = data[(data['DevType']=='Developer, full-stack') & (data['Country'] == 'Pakistan')].shape[0]
pak_devs
```

```python
india_devs = data[(data['DevType']=='Developer, full-stack')&(data['Country'] == 'India')].shape[0]
india_devs
```

```python
afg_dev = data[(data['DevType']=='Developer, full-stack')&(data['Country'] == 'Afghanistan')].shape[0]
afg_dev
```

```python
usa_dev = data[(data['DevType']=='Developer, full-stack')&(data['Country'] == 'United States of America')].shape[0]
usa_dev
```

```python
germ_dev = data[(data['DevType']=='Developer, full-stack')&(data['Country'] == 'Germany')].shape[0]
germ_dev
```

```python
srilanka_dev = data[(data['DevType']=='Developer, full-stack')&(data['Country'] == 'Sri Lanka')].shape[0]
srilanka_dev
```

```python
dev_acros_countries = pd.DataFrame({
    'Country':['Pakistan','India','Afghanistan','USA','Germany','Sri Lanka'],
    'Devs':[pak_devs,india_devs,afg_dev,usa_dev,germ_dev,srilanka_dev]
})
```

```python
plt.bar(dev_acros_countries['Country'],dev_acros_countries['Devs'])
dev_acros_countries = dev_acros_countries.sort_values(by='Devs',ascending=False)
plt.title('Developers Across Countries')
for index,values in enumerate(dev_acros_countries):
    plt.bar_label(plt.gca().containers[0],labels=dev_acros_countries['Devs'])
plt.xlabel('Countries')
plt.ylabel('Devs')
plt.show()
```
